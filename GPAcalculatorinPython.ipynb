```python
# Define the Student class, the base class for the transcript system
class Student:
    def __init__(self, name, reg_number, dob, program):
        # Initialize student attributes
        self.name = name
        self.reg_number = reg_number
        self.dob = dob
        self.program = program

    # Method to input student information
    def read_student_info(self):
        self.name = input("Enter student name: ")
        self.reg_number = input("Enter student registration number: ")
        self.dob = input("Enter student date of birth: ")
        self.program = input("Enter student program: ")

        

    # Method to display student information
    def display_student_info(self):
        print("Student Name:", self.name)
        print("Registration Number:", self.reg_number)
        print("Date of Birth:", self.dob)
        print("Program:", self.program)


# Define the Course class, inheriting from Student
class Course(Student):
    def __init__(self, name, reg_number, dob, program):
        # Initialize course attributes and inherit student attributes
        super().__init__(name, reg_number, dob, program)
        self.course_code = ""
        self.course_title = ""
        self.course_grade = ""

    # Method to input course information
    def read_course_info(self):
        self.course_code = input("Enter course code: ")
        self.course_title = input("Enter course title: ")
        self.course_grade = input("Enter course grade: ")

    # Method to display course information
    def display_course_info(self):
        self.display_student_info()
        print("Course Code:", self.course_code)
        print("Course Title:", self.course_title)
        print("Course Grade:", self.course_grade)


# Define the Semester class, inheriting from Course
class Semester(Course):
    def __init__(self, name, reg_number, dob, program):
        # Initialize semester attributes and inherit course attributes
        super().__init__(name, reg_number, dob, program)
        self.year = ""
        self.semester = ""
        self.courses = []

    # Method to input semester information
    def read_semester_info(self):
        self.year = input("Enter year: ")
        self.semester = input("Enter semester: ")
        num_courses = int(input("Enter the number of courses in this semester: "))
        for i in range(num_courses):
            course = Course(self.name, self.reg_number, self.dob, self.program)
            course.read_course_info()
            self.courses.append(course)

    # Method to compute semester GPA
    def compute_semester_gpa(self):
        total_points = 0
        total_credits = 0
        for course in self.courses:
            if course.course_grade == "A":
                points = 4
            elif course.course_grade == "A-":
                points = 3.7
            elif course.course_grade == "B+":
                points = 3.3
            elif course.course_grade == "B":
                points = 3.0
            elif course.course_grade == "B-":
                points = 2.7
            elif course.course_grade == "C+":
                points = 2.3
            elif course.course_grade == "C":
                points = 2.0
            else:
                points = 0
            total_points += points
            total_credits += 3  # Assuming all courses are 3 credits
        if total_credits == 0:
            return 0
        return total_points / total_credits

    # Method to display semester information
    def display_semester_info(self):
        self.display_student_info()
        print("Year:", self.year)
        print("Semester:", self.semester)
        print("Courses in this semester:")
        for course in self.courses:
            course.display_course_info()
        semester_gpa = self.compute_semester_gpa()
        print("Semester GPA:", semester_gpa)


# Define the Transcript class, inheriting from Semester
class Transcript(Semester):
    def __init__(self, name, reg_number, dob, program):
        # Initialize transcript attributes and inherit semester attributes
        super().__init__(name, reg_number, dob, program)
        self.semesters = []

    # Method to compute cumulative GPA (CGPA)
    def compute_cgpa(self):
        total_gpa = 0
        total_semesters = len(self.semesters)
        for semester in self.semesters:
            total_gpa += semester.compute_semester_gpa()
        if total_semesters == 0:
            return 0
        return total_gpa / total_semesters

    # Method to display the complete transcript
    def display_transcript(self):
        self.display_student_info()
        print("Transcript:")
        for semester in self.semesters:
            semester.display_semester_info()
        cgpa = self.compute_cgpa()
        print("Cumulative GPA (CGPA):", cgpa)


# Example usage:
student = Transcript("", "", "", "")
student.read_student_info()

num_semesters = int(input("Enter the number of semesters: "))
for i in range(num_semesters):
    semester = Semester(student.name, student.reg_number, student.dob, student.program)
    semester.read_semester_info()
    student.semesters.append(semester)

student.display_transcript()
```
